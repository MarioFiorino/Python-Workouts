#Node class 

In doubly linked list  each node consists of 3 component:

1 for the information-data, 

2 to link consecutive nodes.

```python
class Node:

  __slots__ = ("data", "post", "ante")      

  def __init__(self, data):
    self.data = data       #   item : information-data
    self.post = None       #   reference of next node
    self.ante = None       #   reference of the previous node

  def __str__(self):
    return f"Node contains the item : {self.data}"  

  ###EXTRA###
  # Just to check how work "@classmethod" as "Costruttore Alternativo"
  @classmethod                           
  def to_int(cls, one):                  
    if type(one) == float:               # Given in input data in float, instantiates a node whose content is an equivalent integer
      return  cls(int(one))
    elif  type(one) == list:
      return cls(int(sum(i for i in one if type(i)==int )))  # Given in input data list, instantiates a node whose content is an equivalent sum of only integers
    else:
      return cls(one)  
```

# Doubly Linked List class

The last node has a next reference to null. The first node has a previous reference to null. 

Doubly linked lists use the head variable to refer to the first element and the tail variable that refers to the last element of the list 

If the linked list is empty, then both head and tail are null references.

*

It is expected for a linked list to support at least the following high-level operations:

- Initialisation: initialise an empty linked list
- Insert: inserts an element into the list
- Delete: removes and returns the specified position element from the list
- Forward traversal: iterate over all the nodes from head to tail
- Backward traversal: iterate over all the nodes from tail to head
- Count: returns the number of elements in the list
- Quit: destroy all nodes and the list

```python
# Doubly Linked List class
class DLL:    

  def __init__(self):
    self.head = None 
    self.tail = None

  ###Building Iterators for Forward traversal
  
  def __iter__(self):                  
        self.current_node = self.head  # initialization
        return self                    # returns the iterator object itself

  def __next__(self):                  # performs the Forward traversal, implementing the rule for the single hop from one element to another
      if self.current_node != None:
          result = self.current_node
          self.current_node = self.current_node.post
          return result                # return the next item in the sequence
      else:
          raise StopIteration          #  when it reachs the end

  ####

  def print_DDL(self):                 #display the content of the list
   if (self.head == None) :
      print("Doubly Linked List is empty")   
   else:
      n_t = self.head
      while n_t != None:
        if n_t.post == None:
          print(n_t.data)
        else:   
          print(n_t.data,end="  <--->  ")
        n_t = n_t.post
      

  def check_emptyDDL(self):        # jut inform user if DDL is empy ot not
    if (self.head == None) and (self.tail == self.head):
      print("Doubly Linked List is empty")
    else:
      print("Doubly Linked List is NOT empty")  

 
  def front_entry(self,n_data):    # Insert new element at the beginning of the list
                                   # n_data is information-data insert in new node to insert
    new_node = Node(n_data) 
    #new_node.ante = None          
    if self.head is not None: 
        new_node.post = self.head
        self.head.ante = new_node
        self.head = new_node         
    else:                          # If the list is empty, new node is both head and tail
        self.head = new_node
        self.tail = new_node


  def count_DLL(self):             # returns the number of elements in the list
      c = 0
      n_t = self.head
      while n_t != None:
        c += 1
        n_t = n_t.post
      return c                  
  

  ### Decorator : Some check for method: del_spe_from_Head()

  def decor_check(func):           

     def inner(self,position):    
       if self.head == None:
         print("DDL is emply, there is no point in talking about position, the return will be None")
         return 
       
       assert  type(position) == int,"An integer should be entered as position"   # Let pass only integer for position

       if position <= 0:
         print("The position given must be minimum one")
         return
       
       return func(self,position)

     return inner      

  ####

  @decor_check
  def del_spe_from_Head(self,position):  # Delete specified position element in DLL, starting counting from head
                                         # If it exists, return the deleted element, otherwise return None
 
      # Determine the position of node in DDL in Forward traversal. 
      # Note: Head get position 1, the next Node after the Head get position 2, ...
      n_t = self.head
      p = 1
      while (n_t != None) and (p < position):       
        p += 1
        n_t = n_t.post
      
      #print("\t",p)
      #print("\t",n_t)   
   
      if (p == 1) and (self.head == self.tail):     # if there is just one element in DLL
        dt = self.head.data
        self.head = None 
        self.tail = None
        return dt

      if p == (self.count_DLL()):   # oppure :  if self.n_t == self.tail:        
        dt =  self.tail.data
        self.tail = n_t.ante
        n_t.ante.post = None
        return dt

      if n_t == self.head:
        dt = self.head.data
        self.head = n_t.post
        n_t.post.ante = None
        return dt

      if n_t == None:
        print("Was given a position too large compared to the current number of items in DLL.The return will be None")
        print("Suggested to use method .count_DLL()\n")
        return     

      dt = n_t.data  
      n_t.post.ante=n_t.ante       # base case
      n_t.ante.post=n_t.post
      return dt


  def quit_all(self):              # destroy all nodes
    p = self.count_DLL()
    while (self.head != None) :
      self.del_spe_from_Head(p)
      p -= 1
    #del p
    #d = dir()
    #print(d)         
    return 

```

# Doubly Linked List class Backward traversal

Using of the concept of inheritance I created a class that allow iteratation over all the nodes from tail to head (overwriting on _iter and _nexr methods of previous class DLL)

```python
# Doubly Linked List Backward traversal class
class DLL_Backward(DLL):    


  #Building Iterators for Backward traversal
  def __iter__(self):                  
        self.current_node = self.tail  # initialization
        return self                    # returns the iterator object itself

  def __next__(self):                  # performs the Backwardtraversal
      if self.current_node != None:
          result = self.current_node
          self.current_node = self.current_node.ante
          return result                # return the next item in the sequence
      else:
          raise StopIteration          #  when it reachs the end
```
